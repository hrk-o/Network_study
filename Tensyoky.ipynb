## 理想化された情報伝達モデル
転職　第3章　情報の流れのダイナミクス　p.49

```python
import random
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'IPAexGothic'
```

```python
x = 1000 # 集団の人数
b = 2 #はじめに知っている人
s = 0.3 # 話す確率
stime = 50
```

```python
x_flag = [0 for i in range(x)] # 0は知らない
for i in range(b):
    x_flag[i] = 1 # 1は知っている
```

```python
b_time = [] #知っている人
b_time.append([b,0,0,0,x-b])
for time in range(1,stime):
    b_time_t = [0,0,0,0,0] # 知ってる人, 話した人が知っていた, 話した人知らなかった(話した), 話した人知らなかった(話してない), そもそも話す人知らない
    x_flag_t = x_flag
    for xind in x_flag_t:
        if xind == 0:#そもそも話す人知らない
            b_time_t[4] += 1
        else:
            g_x = random.randint(0,x-1)# 話した人フラグ
            if x_flag_t[g_x] == 1:# 話した人が知っていた
                b_time_t[1] += 1
            else :
                if random.random() < s:#話した人知らなかった(話した)
                    x_flag[g_x] = 1
                    b_time_t[2] += 1
                else :#話した人知らなかった(話してない)
                    x_flag[g_x] = 0
                    b_time_t[3] += 1
    b_time_t[0] = sum(x_flag)
    b_time.append(b_time_t)
```

```python
plt.plot(b_time)
plt.legend(["知っている人","話した人が知っていた","話した人知らなかった(話した)","話した人知らなかった(話してない)","そもそも話す人知らない"], fontsize = 8)
plt.show()
```
